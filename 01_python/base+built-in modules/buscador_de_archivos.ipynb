```python
text = '999-999-9999'
numero = re.search('\d{3}-\d{3}-\d{4}',text)
numero.group()
```

```python
def filter(documento,busqueda):
    arch = open(documento,'r')
    texto = arch.read()
    resultado = re.findall(busqueda,texto)
    arch.close
    if len(resultado) > 0:
        print(resultado, "\n", documento)
```

```python
documento = 'C:\\Users\\Alberto\\python notebooks\\unpaked_file\\extracted_content\\Instructions.txt'
busqueda = '\d'
filter(documento,busqueda)
```
