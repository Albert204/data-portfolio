```python
import random
suit = ("hearts", "diamond", "spades", "clubs")
rank = ("ace","two", "three", "four", "five", "six", "seven", "eight", "nine", "ten", "jack", "queen", "king")
value = {"ace":11, "two":2, "three":3, "four":4, "five":5, "six":6, "seven":7, "eight":8, "nine":9, "ten":10, "jack":10,
         "queen":10, "king":10}
```

```python
class Cards:
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit
        self.value = value[rank]
    
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    def __init__(self):
        self.deck = []
        for i in suit:
            for n in rank:
                self.deck.append(Cards(n, i))
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def draw_one(self):
        try:
            return self.deck.pop()
        except IndexError:
            print("out of cards")
```

```python
class Player(Cards):
    
    def __init__(self, account = 1000):
        self.balance = account
        self.player_hand = []
    
    def withdrawl(self, amount):
        self.balance -= amount
        print(f"{amount} withdrawl from account, {self.balance} left")
    
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited into account, {self.balance} left")

    def hand_value(self):
        player_value = 0
        ace_count = 0
        
        for card in self.player_hand:
            if card.rank == "ace":
                ace_count +=1
                player_value += card.value
            else:
                player_value += card.value
                
        while player_value > 21 and ace_count > 0:
            player_value -= 10
            ace_count -= 1
        
        return player_value
    
    def draw_card(self,new_cards):
        if type(new_cards) == type([]):
            self.player_hand.extend(new_cards)
            
        else:
            self.player_hand.append(new_cards)
    
    def __str__(self):
        if not self.player_hand:
            return "Player has no cards"
        
        return "Player has: " + ", ".join(str(card) for card in self.player_hand) + " value: " + str(self.hand_value())
```

```python
class Dealer(Cards):
    
    def __init__(self):
        self.dealer_hand = []
    
    def hand_value(self):
        dealer_value = 0
        ace_count = 0
        
        for card in self.dealer_hand:
            if card.rank == "ace":
                ace_count +=1
                dealer_value += card.value
            else:
                dealer_value += card.value
                
        while dealer_value > 21 and ace_count > 0:
            dealer_value -= 10
            ace_count -= 1
            
        return dealer_value

    def draw_card(self,new_cards):
        if type(new_cards) == type([]):
            self.dealer_hand.extend(new_cards)
            
        else:
            self.dealer_hand.append(new_cards)
    
    def __str__(self):
        if not self.dealer_hand:
            return "Dealer has no cards"
        
        return "Dealer has: " + ", ".join(str(card) for card in self.dealer_hand) + " value: " + str(self.hand_value())
```

```python
def win(player,bet):
    print(f"Player wins, deposited {bet*2}")
    player.deposit(bet*2)
    
def lose(player,bet):
    print(f"Player lost {bet}")
    
def tie(player,bet):
    print(f"Player tied, balance replenished")
    player.deposit(bet)
```
