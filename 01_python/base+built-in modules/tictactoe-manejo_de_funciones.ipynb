```python
def select_player():
    ready = False
    while ready == False:
        player1 = input ("Select O or X to start:")
        if player1.isalpha ():
            player1 = str(player1)
            if player1 != ("O"):
                print ("Select O or X")
            else:
                ready = True
            if player1 != ("X"):
                print ("Select O or X")
            else:
                ready = True
        else:
            print("Is a capital letter...")
            pass
    print ("Starts " + player1)
    return player1
```

```python
def full_board(board):
    counter = 0
    for index in board:
        if index == " ":
            return False
        else:
            counter += 1
    if counter == 9:
        return True
#version simplificada
'''def full_board(board):
    return " " not in board'''
```

```python
def TicTacToe():
    rep = True
    while rep == True:

        board = [" "," "," "," "," "," "," "," "," "]

        turn = 0
        loop = True

        player1 = select_player()
        if player1 == "O":
            player2 = "X"
        else:
            player2 = "O"
        cell(board)


        while loop == True:
            valid_move = False

            print(turn," zero")
            if turn == 0:

                while valid_move == False:
                    position = select_position()
                    if free_position(board, position) == True:
                        make_move(board, player1, position)
                        valid_move = True
                    else:
                        print("invalid position")
                print(turn," first")
                turn = 1
                print(turn," second")

            else:

                while valid_move == False:
                    position = select_position()
                    if free_position(board, position) == True:
                        make_move(board, player2, position)
                        valid_move = True
                    else:
                        print("invalid position")
                print(turn," third")
                turn = 0
                print(turn," fourth")

            if win_check(board) == True:
                if turn == 1:
                    print("Player 1 Won")
                else:
                    print("Player 2 Won")
                loop = False

            if full_board(board) == True:
                print("Draw")
                loop = False
        rep = replay()
```
